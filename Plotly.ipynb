```python
df = pd.read_csv("/content/drive/MyDrive/Estatística do Básico ao Avançado-EBA/Cópia de top10s.csv", encoding = "ISO-8859-1")
```

```python

df.head()
```

#Cores a serem utilizadas:
# '#1db954','#0c4f24','#D3E5D9','#679778','#57615A'

#Gráfico de Barras

```python

top_genre=(df.
groupby('top genre')
['title']
.count()
.to_frame()
.reset_index()
[:15]) #show 15 maiores
```

```python
top_genre
```

```python

fig=px.bar(top_genre,x='top genre',y='title')
fig.update_traces(marker_color='#1db954')
fig.update_layout( title_text= 'Número de músicas por genêro', font_family='Courier New',
                  title_font_family='Courier New',
                   title_font_color='#1db954',
                   plot_bgcolor='#191414' )
fig.update_yaxes(showgrid=True,zeroline=True,visible=False,hoverformat=True)
fig.update_xaxes(showgrid=False,zeroline=True,visible=True,hoverformat=False, categoryorder='total descending' )
fig.show()
```

#Gráfico de Barras empilhadas

```python
df_empilhado=(df.
              groupby(['year','top genre'])
              ['title']
              .count()
              .to_frame()
              .reset_index()
              )
```

```python
df_empilhado.head()
```

```python
fig2=px.bar(df_empilhado,x='year',y='title',color='top genre', barmode ='stack')
fig2.show()
```

```python
df_empilhado2=df_empilhado.loc[df_empilhado['top genre'].isin(['dance pop','pop','canadian pop','barbadian pop', 'boy band'])]
```

```python
fig2=px.bar(df_empilhado2,x='year',y='title',color='top genre', color_discrete_sequence=['#1db954','#0c4f24','#D3E5D9','#679778','#57615A'], barmode ='stack')
fig.update_layout( title_text= 'Número de músicas por genêro', font_family='Courier New',
                  title_font_family='Courier New',
                   title_font_color='#1db954',
                   plot_bgcolor='#807E7E' )
fig.update_yaxes(showgrid=True,zeroline=True,visible=False,hoverformat=True)
fig.update_xaxes(showgrid=False,zeroline=True,visible=True,hoverformat=False, categoryorder='total descending',tickmode='linear' )
fig2.show()
```

#*GRÁFICO De PIZZA e de ROSCA*


```python
figpizza=px.pie(df_empilhado,values='title',names='year')
figpizza.show()
```

```python
df3=(df_empilhado.groupby('year')
['title']
     .sum()
     .to_frame()
     .reset_index()
     .query("year>=2017"))
```

```python
figpizza=px.pie(df3,values='title',names='year',color_discrete_sequence=['#1db954','#0c4f24','#D3E5D9','#679778','#57615A'])
figpizza.show()
```

```python
df3.head()
```

```python
#rosca
figrosca=px.pie(df_empilhado,values='title',names='year')
figrosca.update_traces(hole=0.1)
figrosca.show()
```

```python
figrosca=px.pie(df3,values='title',names='year',color_discrete_sequence=['#1db954','#0c4f24','#D3E5D9','#679778','#57615A'])
figrosca.update_traces(hole=0.5)
figrosca.show()
```
