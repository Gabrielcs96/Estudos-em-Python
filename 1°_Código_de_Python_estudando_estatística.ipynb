#Operações com Matrizes

```python
matriz1= np.array([[14,2,3],[48,6,8]])
```

```python
matriz2=np.array([[1,21,3],[4,54,66]])
```

```python
print(matriz1)
```

```python
print(matriz2)
```

```python
#Soma de matrizes
print(np.add(matriz1,matriz2))
```

```python
#Subtração de matrizes

print(np.subtract(matriz1,matriz2))
```

```python
#Multiplicação
print(np.multiply(matriz1,matriz2))
#ou eu posso usar apenas operadores aritiméticos simples
```

```python
#Calculos diagonais -
```
